```python
def get_position_matrix(data_shape):
  position = np.zeros(data_shape)
  d = position.shape[1]
  for i in range(position.shape[0]):
    for j in range(round(position.shape[1]/2)):
      w = 1 / 10000**(2*j/d)
      position[i, 2*j] = np.sin(w*i)
      position[i, 2*j+1] = np.cos(w*i)
  return position
```

```python
def po_data(data, po_matrix, reduce_rate):
  data_po = data.copy()
  data_po = data_po.astype(np.float32)
  data_length = data_po.shape[0]
  for i in range(data_length):
    data_po[i,:] = data_po[i,:] + po_matrix * reduce_rate
  
  return data_po
```

```python
po_list = []
data_shape = X_train[1,:].shape
po_matrix = get_position_matrix(data_shape)
for data in X_list:
  po_list.append(po_data(data, po_matrix, 1))
```

```python
def get_sailency_map(data_set):
  def loss(output):
    n = output.shape[0]
    con = []
    for i in range(n):
      con.append(output[i][0])
    return tuple(con)

  def model_modifier(m):
    m.layers[-1].activation = tf.keras.activations.linear
    return m

  X = np.asarray(data_set)

  from tensorflow.keras import backend as K
  from tf_keras_vis.saliency import Saliency
  from tf_keras_vis.utils import normalize

  # Create Saliency object.
  # If `clone` is True(default), the `model` will be cloned,
  # so the `model` instance will be NOT modified, but it takes a machine resources.
  saliency = Saliency(model, model_modifier=model_modifier, clone=False)

  # Generate saliency map
  saliency_map = saliency(loss, X)
  saliency_map = normalize(saliency_map)
  #np.save(DATA_PROJECT+'saliency.npy', saliency_map)
  return saliency_map
```

```python
# Render
def plot_saliency_map(plot_data):
  saliency_map = get_sailency_map(plot_data)
  f, ax = plt.subplots(**subplot_args)
  for i, title in enumerate(image_titles):
    ax[i].set_title('Sequence_'+title, fontsize=14)
    ax[i].imshow(saliency_map[i].reshape(1,-1), cmap='jet')
  plt.tight_layout()
  plt.show()
```

```python
image_titles = []
subplot_args = { 'nrows': PLOT_NUM, 'ncols': 1, 'figsize': (10, 10),
          'subplot_kw': {'xticks': [], 'yticks': []} }
for i in range(PLOT_NUM):
  image_titles.append(str(i))

plot_saliency_map(plot_data)
```

```python
def count_genes(count_data, K):

  def find_max_sub(sequence, k):
    sum_list = []
    for i in range(len(sequence)-k):
      sum_list.append(sequence[i:i+k].sum())
    sum_list = np.array(sum_list)
    return sum_list.argmax()


  def onehot_to_gene(one_hot):
    gene_list = ['A','C','G','T']
    n = one_hot.shape[0]
    genes = []
    for i in range(n):
      single_gene = gene_list[one_hot[i].argmax()]
      genes.append(single_gene)
    return genes

  saliency_map = get_sailency_map(count_data)
  n = saliency_map.shape[0]
  data_str = []
  count_dict = {}
  for i in range(n):
    data_str.append(onehot_to_gene(X_train[i,:,:]))
    start_index = find_max_sub(saliency_map[i,:], K)
    sub = data_str[i][start_index:start_index+K]
    sub_str = ''.join(str(e) for e in sub)
    if sub_str not in count_dict.keys():
      count_dict[sub_str] = 1
    else:
      count_dict[sub_str] = count_dict[sub_str]+1

  count_dict = {k: v for k, v in sorted(count_dict.items(), key=lambda item: item[1], reverse=True)}
  return count_dict
```
