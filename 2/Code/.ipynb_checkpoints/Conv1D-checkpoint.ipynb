```python
DATA_PROJECT = '/content/drive/MyDrive/Project/Data/'
SAVE_PATH = '/content/drive/MyDrive/Project/Model/'
EPOCHS = 100

X_train = np.load(DATA_PROJECT+'train.npy')
X_train = X_train.astype(float)
y_train = np.load(DATA_PROJECT+'train_labels.npy')

X_valid = np.load(DATA_PROJECT+'valid.npy')
X_valid = X_valid.astype(float)
y_valid = np.load(DATA_PROJECT+'valid_labels.npy')

X_test = np.load(DATA_PROJECT+'test.npy')
X_test = X_test.astype(float)
y_test = np.load(DATA_PROJECT+'test_labels.npy')

X_list = [X_train, X_valid, X_test]
y_list = [y_train, y_valid, y_test]
```

```python
def get_position_matrix(data_shape):
  position = np.zeros(data_shape)
  d = position.shape[1]
  for i in range(position.shape[0]):
    for j in range(round(position.shape[1]/2)):
      w = 1 / 10000**(2*j/d)
      position[i, 2*j] = np.sin(w*i)
      position[i, 2*j+1] = np.cos(w*i)
  return position
  
```

```python
def po_data(data, po_matrix, reduce_rate):
  data_po = data.copy()
  data_po = data_po.astype(np.float32)
  data_length = data_po.shape[0]
  for i in range(data_length):
    data_po[i,:] = data_po[i,:] + po_matrix * reduce_rate
  
  return data_po
```

```python
po_list = []
data_shape = X_train[1,:].shape
po_matrix = get_position_matrix(data_shape)
for data in X_list:
  po_list.append(po_data(data, po_matrix, 1))
```

```python
from keras.models import Model
from keras.layers import Input
from keras.layers import Activation
from keras.layers import Conv1D
from keras.layers import MaxPooling1D, AveragePooling1D
from keras.layers import add
from keras.layers.merge import concatenate

# function for creating an identity or projection residual module
def residual_module(layer_in, n_filters, size_1=3, size_2=3):
  merge_input = layer_in
  # check if the number of filters needs to be increase, assumes channels last format
  if layer_in.shape[-1] != n_filters:
    merge_input = Conv1D(n_filters, 1, padding='same', activation='relu', kernel_initializer='he_normal')(layer_in)

  conv1 = Conv1D(n_filters, size_1, padding='same', activation='relu', kernel_initializer='he_normal')(layer_in)
  conv2 = Conv1D(n_filters, size_2, padding='same', activation='linear', kernel_initializer='he_normal')(conv1)
  #merge_input = BatchNormalization()(merge_input)
  #merge_input = Activation('relu')(merge_input)
  layer_out = add([conv2, merge_input])
  layer_out = Activation('relu')(layer_out)
  return layer_out

base_filter_num = 32
visible = Input(shape=(101, 4))
initial = Conv1D(base_filter_num, 5, padding='same', strides=1)(visible)
block_1 = residual_module(initial, base_filter_num)
block_2 = residual_module(block_1, base_filter_num)
block_3 = residual_module(block_2, base_filter_num)
pooling_1 = MaxPooling1D(2, 2)(block_3)

block_4 = residual_module(pooling_1, base_filter_num*2)
block_5 = residual_module(block_4, base_filter_num*2)
block_6 = residual_module(block_5, base_filter_num*2)
pooling_2 = MaxPooling1D(2, 2)(block_6)

block_7 = residual_module(pooling_2, base_filter_num*4)
block_8 = residual_module(block_7, base_filter_num*4)
block_9 = residual_module(block_8, base_filter_num*4)
pooling_3 = MaxPooling1D(2, 2)(block_9)

block_10 = residual_module(pooling_3, base_filter_num*8)
block_11 = residual_module(block_10, base_filter_num*8)
block_12 = residual_module(block_11, base_filter_num*8)
pooling_4 = MaxPooling1D(2, 2)(block_12)

block_13 = residual_module(pooling_4, base_filter_num*16)
block_14 = residual_module(block_13, base_filter_num*16)
block_15 = residual_module(block_14, base_filter_num*16)
pooling_5 = AveragePooling1D(2, 1)(block_15)

flatten = Flatten()(pooling_5)
dropout_1 = Dropout(0.5)(flatten)
output = Dense(1, activation='sigmoid', kernel_regularizer='l1_l2')(dropout_1)

model_res = Model(inputs=visible, outputs=output)
model_res.compile(optimizer='adam', loss='BinaryCrossentropy', metrics=['accuracy', 'AUC'])
model_res.summary()
# plot model architecture
#plot_model(model_res, show_shapes=True, to_file='residual_module.png')
```

```python
checkpoint_filepath = SAVE_PATH + 'Resnet'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath = checkpoint_filepath,
    save_weights_only = False,
    monitor = 'val_accuracy',
    mode = 'max',
    save_best_only = True)
csv_logger = CSVLogger(checkpoint_filepath+'/train_history.csv', append=False)
```

```python
model_res.fit(po_list[0], y_train, epochs=EPOCHS, batch_size=128, validation_data=(po_list[1], y_valid),
        verbose=2, callbacks=[model_checkpoint_callback, csv_logger])
```

**Train the VGG model**

```python
model_1d = Sequential([
    layers.Conv1D(32, 3, input_shape=(101,4), activation='relu', use_bias=True, padding='same'),
    layers.Conv1D(32, 3, activation='relu', use_bias=True, padding='same'),
    layers.MaxPool1D(2, 2),

    layers.Conv1D(64, 3, activation='relu', use_bias=True, padding='same'),
    layers.Conv1D(64, 3, activation='relu', use_bias=True, padding='same'),
    layers.MaxPool1D(2, 2),

    layers.Conv1D(128, 3, activation='relu', use_bias=True, padding='same'),
    layers.Conv1D(128, 3, activation='relu', use_bias=True, padding='same'),
    layers.MaxPool1D(2, 2),

    Flatten(),
    Dense(1024, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid', kernel_regularizer='l1_l2')
                       
])
model_1d.compile(optimizer='adam', loss='BinaryCrossentropy', metrics=['accuracy', 'AUC'])
model_1d.summary()
```

```python
checkpoint_filepath = SAVE_PATH + 'VGG'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath = checkpoint_filepath,
    save_weights_only = False,
    monitor = 'val_accuracy',
    mode = 'max',
    save_best_only = True)
csv_logger = CSVLogger(checkpoint_filepath+'/train_history.csv', append=False)
```

```python
model_1d.fit(po_list[0], y_train, epochs=EPOCHS, batch_size=128, validation_data=(po_list[1], y_valid),
        verbose=2, callbacks=[model_checkpoint_callback, csv_logger])
```

```python
# function for creating a naive inception block
def inception_module(layer_in, f1, f2, f3, f4, f5):
  conv1 = Conv1D(f1, 3, padding='same', activation='relu')(layer_in)
  conv3 = Conv1D(f2, 5, padding='same', activation='relu')(layer_in)
  conv5 = Conv1D(f3, 7, padding='same', activation='relu')(layer_in)
  conv10 = Conv1D(f4, 10, padding='same', strides=2, activation='relu', use_bias=True)(layer_in)
  conv20 = Conv1D(f5, 20, padding='same', strides=2, activation='relu', use_bias=True)(layer_in)
  pool = MaxPooling1D(3, strides=1, padding='same')(layer_in)
  layer_out = concatenate([conv1, conv3, conv5, pool], axis=-1)
  return layer_out


base_filter_num = 16
visible = Input(shape=(101,4))
layer_1 = inception_module(visible, 128, 64, 32, 16, 16)
layer_2 = inception_module(layer_1, 128, 64, 32, 16, 16)
layer_3 = Conv1D(64, 3, padding='same', activation='relu')(layer_2)
flatten = Flatten()(layer_3)
#norm = BatchNormalization()(flatten)
dropout_1 = Dropout(0.5)(flatten)
output = Dense(1, activation='sigmoid', kernel_regularizer='l1_l2')(dropout_1)

model_incept = Model(inputs=visible, outputs=output)
model_incept.compile(optimizer='adam', loss='BinaryCrossentropy', metrics=['accuracy', 'AUC'])
model_incept.summary()
# plot model architecture
#plot_model(model_res, show_shapes=True, to_file='residual_module.png')
```

```python
checkpoint_filepath = SAVE_PATH + 'Inception'
model_checkpoint_callback = tf.keras.callbacks.ModelCheckpoint(
    filepath = checkpoint_filepath,
    save_weights_only = False,
    monitor = 'val_accuracy',
    mode = 'max',
    save_best_only = True)
csv_logger = CSVLogger(checkpoint_filepath+'/train_history.csv', append=False)
```

```python
model_incept.fit(po_list[0], y_train, epochs=EPOCHS, batch_size=128, validation_data=(po_list[1], y_valid),
        verbose=2, callbacks=[model_checkpoint_callback, csv_logger])
```
